```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd
```

# Question 1

### load in iris data

```python
data = pd.read_csv('https://tinyurl.com/ChrisCoDV/iris.csv')
data
```

# Question 2

### 3 seperate box plots for each variety showing distribution of the 4 measurement types

```python
selected = ['sepal.length' , 'sepal.width' , 'petal.length' , 'petal.width']


varieties = ['Setosa','Versicolor','Virginica']

for variety in varieties:
    data_variety = data[data['variety']== variety]
   # data_versicolor = data[data['variety']=='Versicolor']
   #data_virginica = data[data['variety']=='Virginica']

    plt.figure(figsize = (8,8))
    plt.boxplot(data_variety[selected], labels=selected)
    plt.xlabel('Variety', fontsize=18)
    plt.ylabel('Measurement (CM)', fontsize=18)
    plt.title(variety + ' distributions', fontsize=20)
    plt.show()
```

# Question 3

### 3 histograms for each variety showing distribution of the 4 measurement types all superimposed on the same plot

```python
x_min = 0
x_max = 8
bin_width = 0.1
n_bins = int((bin_width + x_max - x_min) / bin_width)
print(str(n_bins) + ' bins')
bins = [(x_min + x * (bin_width + x_max - x_min)/n_bins) for x in range(int(n_bins))]

for variety in varieties:
    data_variety = data[data['variety']== variety]
    plt.figure(figsize = (8,8))
    for label in selected:
        plt.hist(data_variety[label], bins, alpha=0.5, label=label)
    plt.title('Iris ' + variety + ' measurement distributions', fontsize=20)
    plt.ylim(ymax=30)
    plt.legend(loc=1)
    plt.show()
```

smallest spread is setosa which confirms the boxplot
