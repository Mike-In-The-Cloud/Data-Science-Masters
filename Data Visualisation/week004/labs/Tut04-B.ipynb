# Question 1

## read in data

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
data = pd.read_csv('https://tinyurl.com/ChrisCoDV/iris.csv')
```

```python
print(data.head())
```

# Quesiton 2

# single vis scatter  sub with all pairs

```python
selected = ['sepal.length', 'sepal.width', 'petal.length',  'petal.width']


counter = 1
fig = plt.figure(figsize=(8,8))
fig.suptitle('Varierty correlations', fontsize=14, position=(0.5,1.0))
for i, name_i in enumerate(selected):
    for j in range(i+1, len(selected)):
        name_j = selected[j]
        sub = fig.add_subplot(3,3, counter)
        sub.set_title(name_i + ' vs ' + name_j, fontsize=10)
        sub.scatter(data[name_i], data[name_j], s=0.5)
        counter += 1
plt.subplots_adjust(wspace = 0.5, hspace = 0.5)
plt.tight_layout()
plt.show()
```

# Question 3

### heatmap showing correlations

```python
plt.figure(figsize=(10,10))
corr = data[selected].corr()
ax = sns.heatmap(corr, vmin=-1, vmax = 1, center = 0, cmap=sns.diverging_palette(220,20,n=200),
                 square=True,
                 annot=True,
                 annot_kws={"size": 8})
ax.set_xticklabels(ax.get_xticklabels(), rotation=45, horizontalalignment='right')
plt.show()
```

# Question 6

### repeat for Versicolor and for Virginica

```python
data_versicolor = data[data['variety']=='Versicolor']
data_virginica = data[data['variety']=='Virginica']
```

```python
def scatterplots(data1,datatitle):
    counter = 1
    fig = plt.figure(figsize=(8,8))
    fig.suptitle(datatitle + ' correlations', fontsize=14, position=(0.5,1.0))
    for i, name_i in enumerate(selected):
        for j in range(i+1, len(selected)):
            name_j = selected[j]
            sub = fig.add_subplot(3,3, counter)
            sub.set_title(name_i + ' vs ' + name_j, fontsize=10)
            sub.scatter(data1[name_i], data1[name_j], s=0.5)
            counter += 1
    plt.subplots_adjust(wspace = 0.5, hspace = 0.5)
    plt.tight_layout()
    plt.show()
```

```python
scatterplots(data_versicolor, "Versicolor")
```

```python
scatterplots(data_virginica, "Virginica")
```

```python
def plotHeatMaps(data1):
    plt.figure(figsize=(10,10))
    corr = data1[selected].corr()
    ax = sns.heatmap(corr, vmin=-1, vmax = 1, center = 0, cmap=sns.diverging_palette(220,20,n=200),
                     square=True,
                     annot=True,
                     annot_kws={"size": 8})
    ax.set_xticklabels(ax.get_xticklabels(), rotation=45, horizontalalignment='right')
    plt.show()
```

```python
plotHeatMaps(data_versicolor)
```

```python
plotHeatMaps(data_virginica)
```
